### Lectura y filtración 

````python
# Prueba de carga del archivo 
cargar_data_csv(DATA_PREP_DIRECTORY + FILE_NAME)
````

````python
###  Leemos el .csv con Spark y esquema 

# Definimos el esquema (tomado de la sugerencia del eda.ipynb)
schema_ventas =  StructType([
    StructField('fecha', DateType(), True),
    StructField('region', StringType(), True),
    StructField('producto', StringType(), True),
    StructField('ventas', IntegerType(), True),
    StructField('clientes', IntegerType(), True)
])

# Leemos el archivo csv 
df = cargar_csv_con_schema(spark, ruta_archivo=DATA_PREP_DIRECTORY + FILE_NAME, schema=schema_ventas)
````

````python
# Ingenieria básica de variables 
df = agregar_columnas_temporales(df, "fecha")
````

````python
# Filtramos datos para 2024
df_filtrado = df.filter(F.col("anio") == 2024)
````

````python
# Agregacion de ventas por region y producto
df_ventas_agg_filtrado = (df_filtrado.select("region","producto","ventas")
                                     .groupBy("region","producto")
                                     .agg(F.sum("ventas").alias("ventas_totales"))
                         )
````

````python
# Agregacion de clientes por region y producto
df_clientes_agg_filtrado = (df_filtrado.select("region","producto","clientes")
                                     .groupBy("region","producto")
                                     .agg(F.sum("clientes").alias("clientes_totales"))
                         )
````

````python
# Producto con mayor venta en cada región

# Definimos la ventana de ventas totales 
window_ventas = Window.partitionBy("region",).orderBy(F.desc("ventas_totales"))

# Filtramos los productos de mayor venta
df_top_productos = (df_ventas_agg_filtrado.withColumn("rank", F.row_number().over(window_ventas)) 
               .filter(F.col("rank") == 1) 
               .drop("rank")
                )
````

````python
#Definimos cuotas de Mercado 

# Totales
total_ventas = df_ventas_agg_filtrado.agg(F.sum("ventas_totales").alias("total")).collect()[0]["total"]
total_clientes = df_clientes_agg_filtrado.agg(F.sum("clientes_totales").alias("total")).collect()[0]["total"]

# Cuotas
df_ventas_agg_filtrado = df_ventas_agg_filtrado.withColumn("ventas_cuota", F.round(F.col("ventas_totales") / F.lit(total_ventas),6))
df_clientes_agg_filtrado = df_clientes_agg_filtrado.withColumn("clientes_cuota", F.round(F.col("clientes_totales") / F.lit(total_clientes),6))
````

````python
# Guardamos los datos en un archivo csv para futura explotacion o visualización

df.toPandas().to_csv(DATA_CLEAN_DIRECTORY + "data_clean.csv", index = False)
df_filtrado.toPandas().to_csv(DATA_CLEAN_DIRECTORY + "data_filtrada_clean.csv", index = False)
df_top_productos.toPandas().to_csv(DATA_CLEAN_DIRECTORY + "data_top_productos.csv", index = False)
df_ventas_agg_filtrado.toPandas().to_csv(DATA_CLEAN_DIRECTORY + "data_agg_ventas.csv", index = False)
df_clientes_agg_filtrado.toPandas().to_csv(DATA_CLEAN_DIRECTORY + "data_agg_clientes.csv", index = False)
````

### Consultas SQL 

**Total de ventas y clientes por región en 2024**

Query: 

``` SQL
-- Consulta para sumar clientes y ventas por región en el año 2024
SELECT
    region,
    SUM(clientes) AS clientes_total,
    SUM(ventas) AS ventas_total
FROM VENTAS
WHERE anio = 2023
GROUP BY region
```

**Top 3 productos con mayores ventas globales en 2024**

Query: 

``` SQL
-- Consulta  top 3 productos más vendidos en 2024
SELECT
    producto,
    SUM(ventas) AS ventas_total
FROM VENTAS
WHERE anio = 2023
GROUP BY producto
ORDER BY ventas_total DESC
LIMIT 3

```

**Consulta del porcentaje de crecimiento de ventas por producto entre 2023 y 2024**

Query: 

``` SQL
-- Consulta del crecimiento porcentual de ventas por producto entre 2023 y 2024

-- CTE: Ventas por producto en 2023
WITH agg_ventas_2023 AS (
    SELECT
        producto,
        SUM(ventas) AS ventas_2023
    FROM VENTAS
    WHERE anio = 2023
    GROUP BY producto
),

-- CTE: Ventas por producto en 2024
agg_ventas_2024 AS (
    SELECT
        producto,
        SUM(ventas) AS ventas_2024
    FROM VENTAS
    WHERE anio = 2024
    GROUP BY producto
)

-- Join externo para comparar ventas año contra año
SELECT
    COALESCE(v24.producto, v23.producto) AS producto,
    COALESCE(ventas_2024, 0) AS ventas_2024,
    COALESCE(ventas_2023, 0) AS ventas_2023,
    -- Variación relativa de ventas: (2024 - 2023) / 2023
    CASE
        WHEN ventas_2023 > 0 THEN (ventas_2024 / ventas_2023) - 1
        WHEN ventas_2024 > 0 THEN 1.0
        ELSE 0.0
    END AS crecimiento_ventas
FROM agg_ventas_2023 v23
FULL OUTER JOIN agg_ventas_2024 v24
    ON v23.producto = v24.producto
```

### Modelo de ML: KMedias (Segméntación)

````python
df_m = df.groupBy("producto", "region").agg(F.mean("ventas").alias("ventas"),F.mean("clientes").alias("clientes") )
````

````python
from pyspark.ml.clustering import KMeans
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.sql import SparkSession

# Armar el vector de características
assembler = VectorAssembler(inputCols=["ventas", "clientes"], outputCol="features")
df_vect = assembler.transform(df)

# Normalizar con StandardScaler (sin cambiar el df original)
scaler = StandardScaler(inputCol="features", outputCol="scaled_features", withMean=True, withStd=True)
scaler_model = scaler.fit(df_vect)
df_scaled = scaler_model.transform(df_vect)

#  Ejecutar KMeans sobre las columnas normalizadas
k = 4
kmeans = KMeans(featuresCol="scaled_features", predictionCol="cluster", k=k, seed=42)
modelo = kmeans.fit(df_scaled)

# Predecir los clusters
df_clustered = modelo.transform(df_scaled)

# Ver resultados
df_clustered.select("ventas", "clientes", "cluster").show(5)
````

````python
df.join(df_clustered.select("ventas", "clientes", "cluster"), [df.ventas == df_clustered.ventas, 
                      df.clientes == df_clustered.clientes], "left").toPandas().to_csv(DATA_CLEAN_DIRECTORY + "clusters.csv")
````

````python
# Convertir a Pandas para graficar
df_pd = df_clustered.select("ventas", "clientes", "cluster").toPandas()

# Graficar
fig,ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(data=df_pd, x="ventas",y= "clientes", hue="cluster", palette="Blues")
ax.set_title(f"KMeans Clustering (k = {k})")
fig.show()
````
